```python
import torch
import einops as einops
import numpy as np
import matplotlib.pyplot as plt
from PIL.Image import fromarray


def display_img(img,size=(1,1)):
    plt.figure(figsize=size)
    plt.imshow(img)
    plt.axis("off")
```

```python
ims = np.load("./test_images.npy", allow_pickle=False)
# There are 6 images of shape 96x96 with 3 color channels packed into tensor
print(ims.shape, ims.dtype)
```

```python
display_img(ims[0])
```

```python
transpose=einops.rearrange(ims[0],"h w c -> w h c")
display_img(transpose)
```

```python
axis_compo=einops.rearrange(ims,"b h w c -> (b h) w c")
display_img(axis_compo,size=(3,3))
```

```python
axis_compo=einops.rearrange(ims,"b h w c -> h (b w) c")
display_img(axis_compo,size=(3,3))
```

```python
axis_compo=einops.rearrange(ims,"(b1 b2) h w c -> (b1 h) (b2 w) c",b1=3)
display_img(axis_compo,size=(3,3))
```

```python
axis_compo=einops.rearrange(ims,"(b1 b2) h w c -> (b1 h) (b2 w) c",b2=3)
display_img(axis_compo,size=(3,3))
```

```python
axis_compo=einops.rearrange(ims,"b (h h1) w c -> (h) (b w h1) c",h1=2)
display_img(axis_compo,size=(3,3))
```

```python
axis_compo=einops.rearrange(ims,"b h w c -> h (w b) c")
display_img(axis_compo,size=(10,10))
```

```python
axis_compo=einops.rearrange(ims,"(b1 b2) h w c -> h (b2 b1 w) c",b1=3)
display_img(axis_compo,size=(10,10))
```

```python
axis_compo=einops.reduce(ims,"b h w c -> h w c","mean")
print(axis_compo.shape)
display_img(axis_compo,size=(10,10))
```
